```python
import pandas as pd
import glob
files = glob.glob("*.csv")
df_list = []
for file in files:
    temp_df = pd.read_csv(file)
    df_list.append(temp_df)
final_df = pd.concat(df_list, ignore_index=True)
final_df = final_df.drop_duplicates()
final_df = final_df.dropna()
final_df.info()
```

```python
print("Initial shape:", final_df.shape)

print("\nData types:")
print(final_df.dtypes)

print("\nMissing values per column:")
print(final_df.isnull().sum())

print("\nDuplicate rows:", final_df.duplicated().sum())

print("\nFinal info:")
final_df.info()
```

```python
final_df.to_csv("cleaned_aadhar_data.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Total activity column banao (age groups ko jod ke)
final_df['total_activity'] = final_df['demo_age_5_17'] + final_df['demo_age_17_']

# Step 2: State-wise aggregation
state_counts = final_df.groupby('state')['total_activity'].sum().sort_values(ascending=False)

# Step 3: Top 10 states
top10_states = state_counts.head(10)

print("Top 10 States by Total Activity:")
print(top10_states)

# -------------------------------
# Bar Chart  Top 10 States
# -------------------------------
plt.figure()
top10_states.plot(kind='bar')
plt.title("Top 10 States by Demographic Activity")
plt.xlabel("State")
plt.ylabel("Total Activity")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# Pie Chart  Share of Top 10 States
# -------------------------------
plt.figure()
top10_states.plot(kind='pie', autopct='%1.1f%%')
plt.title("Share of Top 10 States in Total Activity")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Date ko proper datetime me convert karo (agar pehle se nahi hai)
final_df['date'] = pd.to_datetime(final_df['date'], errors='coerce')

# Step 2: Total activity (agar pehle se nahi banaya)
if 'total_activity' not in final_df.columns:
    final_df['total_activity'] = final_df['demo_age_5_17'] + final_df['demo_age_17_']

# Step 3: Year & Month columns banao
final_df['year'] = final_df['date'].dt.year
final_df['month'] = final_df['date'].dt.month

# Year-wise Trend
yearly_trend = final_df.groupby('year')['total_activity'].sum()

plt.figure()
plt.plot(yearly_trend.index, yearly_trend.values)
plt.title("Year-wise Demographic Activity Trend")
plt.xlabel("Year")
plt.ylabel("Total Activity")
plt.show()

#  Monthly Trend (Spikes)
monthly_trend = final_df.groupby('month')['total_activity'].sum()

plt.figure()
plt.plot(monthly_trend.index, monthly_trend.values)
plt.title("Monthly Demographic Activity Spikes")
plt.xlabel("Month")
plt.ylabel("Total Activity")
plt.show()

# Print values for reference
print("Year-wise Activity:")
print(yearly_trend)

print("\nMonthly Activity:")
print(monthly_trend)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Safety: Date & total_activity ensure
final_df['date'] = pd.to_datetime(final_df['date'], errors='coerce')
final_df['total_activity'] = final_df['demo_age_5_17'] + final_df['demo_age_17_']

# DISTRICT HOTSPOTS
district_counts = final_df.groupby('district')['total_activity'].sum().sort_values(ascending=False)

top_districts = district_counts.head(20)

plt.figure()
plt.bar(range(len(top_districts)), top_districts.values)
plt.xticks(range(len(top_districts)), top_districts.index, rotation=90)
plt.title("Top 20 District Hotspots by Activity")
plt.xlabel("District")
plt.ylabel("Total Activity")
plt.show()

# -------------------------------
# PINCODE HOTSPOTS
# -------------------------------
pincode_counts = final_df.groupby('pincode')['total_activity'].sum().sort_values(ascending=False)

top_pincodes = pincode_counts.head(20)

plt.figure()
plt.bar(range(len(top_pincodes)), top_pincodes.values)
plt.xticks(range(len(top_pincodes)), top_pincodes.index, rotation=90)
plt.title("Top 20 Pincode Hotspots by Activity")
plt.xlabel("Pincode")
plt.ylabel("Total Activity")
plt.show()

#HEATMAP (District vs Activity)
heatmap_data = np.array(top_districts.values).reshape(-1, 1)

plt.figure()
plt.imshow(heatmap_data)
plt.yticks(range(len(top_districts)), top_districts.index)
plt.xticks([0], ["Activity"])
plt.title("District-Level Hotspot Heatmap (Top 20)")
plt.show()

# Print values
print("Top 20 District Hotspots:")
print(top_districts)

print("\nTop 20 Pincode Hotspots:")
print(top_pincodes)
```
