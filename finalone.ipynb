```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("merged_clean_data.csv", dtype=str)

df["date"] = pd.to_datetime(df["date"], errors="coerce")

df["age_0_5"] = pd.to_numeric(df["age_0_5"], errors="coerce")
df["age_5_17"] = pd.to_numeric(df["age_5_17"], errors="coerce")
df["age_18_greater"] = pd.to_numeric(df["age_18_greater"], errors="coerce")

df["total_enrolment"] = (
    df["age_0_5"] +
    df["age_5_17"] +
    df["age_18_greater"]
)


df["Year"] = df["date"].dt.year

yearwise_enrolment = df.groupby("Year")["total_enrolment"].sum()

plt.figure()
plt.plot(yearwise_enrolment.index, yearwise_enrolment.values, marker='o')
plt.xlabel("Year")
plt.ylabel("Total Enrolment")
plt.title("Year-wise Enrolment Trend")
plt.grid(True)
plt.show()

print("✅ Year-wise enrolment graph generated successfully")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("merged_clean_data.csv", dtype=str)

df["date"] = pd.to_datetime(df["date"], errors="coerce")

df["age_0_5"] = pd.to_numeric(df["age_0_5"], errors="coerce")
df["age_5_17"] = pd.to_numeric(df["age_5_17"], errors="coerce")
df["age_18_greater"] = pd.to_numeric(df["age_18_greater"], errors="coerce")

df["total_enrolment"] = (
    df["age_0_5"] +
    df["age_5_17"] +
    df["age_18_greater"]
)

df["Month"] = df["date"].dt.month
df["Month_Name"] = df["date"].dt.month_name()

monthwise_enrolment = df.groupby("Month_Name")["total_enrolment"].sum()

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]
monthwise_enrolment = monthwise_enrolment.reindex(month_order)

plt.figure()
plt.plot(monthwise_enrolment.index, monthwise_enrolment.values, marker='o')
plt.xlabel("Month")
plt.ylabel("Total Enrolment")
plt.title("Month-wise Enrolment Trend")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

print("✅ Month-wise enrolment line graph created successfully")
```

```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("merged_clean_data.csv", dtype=str)

df["age_0_5"] = pd.to_numeric(df["age_0_5"], errors="coerce")
df["age_5_17"] = pd.to_numeric(df["age_5_17"], errors="coerce")
df["age_18_greater"] = pd.to_numeric(df["age_18_greater"], errors="coerce")

age_group_totals = {
    "Age 0–5": df["age_0_5"].sum(),
    "Age 5–17": df["age_5_17"].sum(),
    "Age 18+": df["age_18_greater"].sum()
}

age_group_series = pd.Series(age_group_totals)

plt.figure()
age_group_series.plot(kind="bar")
plt.xlabel("Age Group")
plt.ylabel("Total Enrolment")
plt.title("Age Group Distribution of Enrolment")
plt.grid(axis="y")
plt.show()

print("✅ Age group distribution analysis graph created successfully")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("merged_clean_data.csv", dtype=str)

df["age_0_5"] = pd.to_numeric(df["age_0_5"], errors="coerce")
df["age_5_17"] = pd.to_numeric(df["age_5_17"], errors="coerce")
df["age_18_greater"] = pd.to_numeric(df["age_18_greater"], errors="coerce")

df["total_enrolment"] = (
    df["age_0_5"] +
    df["age_5_17"] +
    df["age_18_greater"]
)

top_10_states = (
    df.groupby("state")["total_enrolment"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure()
top_10_states.plot(kind="bar")
plt.xlabel("State")
plt.ylabel("Total Enrolment")
plt.title("Top 10 States by Total Enrolment")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y")
plt.show()

print("✅ Top 10 regions (states) visualization created successfully")
```
